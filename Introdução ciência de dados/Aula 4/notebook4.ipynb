```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report


np.random.seed(42) # define the seed (important to reproduce the results)


data = pd.read_csv(r'G:\Meu Drive\MBA_data_science\Introdução ciência de dados\Aula 4\data\vertebralcolumn-2C.csv', header=(0))
data = data.dropna(axis='rows') #remove NaN


data = data.to_numpy()
nrow,ncol = data.shape
y = data[:,-1]
X = data[:,0:ncol-1]


scaler = StandardScaler().fit(X)
X = scaler.transform(X)


from sklearn.model_selection import train_test_split
p = 0.2 # fraction of elements in the test set
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = p, random_state = 42, train_size=0.8)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

y_pred = gnb.fit(x_train, y_train).predict(x_test)
```

```python
gnb.score(X=x_test, y=y_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(f"Number of mislabeled points out of a total {x_test.shape[0]} points : {(y_test != y_pred).sum()}")
```
