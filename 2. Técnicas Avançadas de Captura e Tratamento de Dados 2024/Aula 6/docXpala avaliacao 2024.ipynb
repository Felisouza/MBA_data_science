### Exercício 1)
Armazene os documentos disponíveis no diretório `DocCol2` em um dicionário onde a chave é o nome do arquivo e o valor é a string contida no arquivo. O documento contendo a menor string é:

a) gr7<br>
b) au1 <br>
c) ch37 <br>
d) au81


````python
folder = "../content/DocCol2/*" # aqui vc deve mudar para o caminho onde o diretório DocCol2 está
files = glob.glob(folder)
````

````python
docs = {}
for fname in files:
    with open(fname,'r') as f:
        key = fname.split(os.sep)[-1]
        docs[key] = f.read()

print(docs.keys())
````

````python
dicionario = {}
for k, v in docs.items():
  dicionario[k] = len(docs[k])

min(dicionario.values())
````

````python
# Encontrar a chave do menor valor
chave_menor_valor = min(dicionario, key=dicionario.get)

# Encontrar o menor valor associado a essa chave
menor_valor = dicionario[chave_menor_valor]

print(f"A chave do menor valor é: {chave_menor_valor}")
print(f"O menor valor é: {menor_valor}")
````

````python
dicionario = {}
lista = []
for k, v in words.items():
  for w in v:
    if w.isalpha() and len(w) >= 2:
      lista.append(w.lower())
      dicionario[k] = lista

#max(dicionario.values())
````

````python
# Encontrar a chave do menor valor
chave_menor_valor = max(dicionario, key=dicionario.get)

# Encontrar o menor valor associado a essa chave
menor_valor = dicionario[chave_menor_valor]

print(f"A chave do menor valor é: {chave_menor_valor}")
print(f"O maior valor é: {len(menor_valor)}")
````

### Exercício 3)
As listas de palavras do dicionário `docsXwords` possuem palavras repetidas. Construa um dicionário chamado `corpus` onde a chave é uma palavra que aparece nas listas de `docsXwords` e os valores correspondem ao número de vezes que a palavra apareceu em `docsXwords`. Por exemplo, supondo que `docsXwords` seja:

```python
'ch3':['the', 'nicen', 'creed']
'au14':['the', 'river', 'creed', 'but']
'gr8':['river', 'the', 'nicen', 'creed']
```
o dicionário `corpus` será:
```python
'the':3
'nicen':2
'creed':3
'river':2
'but':1
```
Quantas palavras estão associadas ao valor 1 no dicionário `corpus`?

a) 2103<br>
b) 1867<br>
c) 2862 <br>
d) 3001

````python
from collections import defaultdict

# Dicionário para contar as ocorrências
contagem_palavras = defaultdict(int)

# Contar as palavras
for lista in docsXwords.values():
    for palavra in lista:
        contagem_palavras[palavra] += 1

# Exibir a contagem
for palavra, contagem in contagem_palavras.items():
    print(palavra)
    print(contagem)
    break
    #print(f"'{palavra}': {contagem}")
````
