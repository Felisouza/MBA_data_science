```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados de uma distribuição normal
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Plotando o histograma usando Seaborn
sns.histplot(data, kde=True, color='skyblue', edgecolor='black')
plt.title('Histograma de uma Distribuição Normal')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Gerando dados de uma distribuição normal
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculando média, moda e mediana
mean = np.mean(data)
mode = stats.mode(data).mode
median = np.median(data)

# Plotando o histograma usando Seaborn
sns.histplot(data, kde=True, color='skyblue', edgecolor='black')

# Adicionando linhas e textos para média, moda e mediana
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label=f'Média = {mean:.2f}')
plt.axvline(mode, color='green', linestyle='dashed', linewidth=1, label=f'Moda = {mode:.2f}')
plt.axvline(median, color='blue', linestyle='dashed', linewidth=1, label=f'Mediana = {median:.2f}')
plt.legend()

plt.title('Histograma de uma Distribuição Normal')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.show()
```
