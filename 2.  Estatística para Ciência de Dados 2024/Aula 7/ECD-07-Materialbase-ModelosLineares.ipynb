# Modelos de regresso

**Objetivos**

Predizer $Y$ a partir do conhecimento de variveis em $X = x$.


Em notao matricial, um modelo linear geral  dado por

$$\large{Y = f(X,\beta)+\epsilon,}$$

em que

- Y  a **varivel resposta** (vetor de variveis aleatrias observveis),

- X contm **variveis preditoras** (matriz conhecida, ou seja, no-aleatria),

- $\beta$  um **vetor de parmetros de interesse**, que queremos estimar,

- $f$  uma funo das variveis preditoras e dos parmetros de interesse,

- $\epsilon$  o **erro aleatrio** (vetor de erros aleatrios no observveis).




# Modelos lineares


**Objetivos**

Predizer $Y$ a partir do conhecimento de variveis em $X = x$ utilizando uma funo linear dos parmetros $\beta$.


Em notao matricial, um modelo linear geral  dado por

$$\large{Y = X\beta+\epsilon,}$$

em que

- Y  a **varivel resposta** (vetor de variveis aleatrias observveis),

- X contm **variveis preditoras** (matriz conhecida, ou seja, no-aleatria),

- $\beta$  um **vetor de parmetros de interesse**, que queremos estimar,

- $\epsilon$  o **erro aleatrio** (vetor de erros aleatrios no observveis).





## Modelo de regresso linear simples


Suponha que so observados os pares $(X_1,Y_1), \ldots, (X_n, Y_n)$ de varivel preditora e resposta, respectivamente.

Um **modelo linear simples** para explicar a variabilidade de $Y$ usando a variabilidade de $X$ seria:

$$\large Y_j=\beta_0 + \beta_1 X_j +\epsilon_j, \mbox{ para } j=1,\ldots,n.$$


Nomenclatura:
-  $Y_j$: $j$-sima observao da varivel resposta (dependente, aleatria),
-  $\beta_0$ e $\beta_1$: parmetros desconhecidos e que queremos estimar (fixo e desconhecido),
-  $X_j$: $j$-sima observao da varivel preditora (fixa, ou seja, no aleatria),
-  $\epsilon_j$: $j$-simo erro aleatrio no observvel.


Suposies:

-  $E(\epsilon_j)=0$ para $j=1,\ldots,n$,
-  $Var(\epsilon_j) = \sigma^2$ para $j=1,\ldots,n$,
-  $Cov(\epsilon_i,\epsilon_j)=0$ para $i,j=1,\ldots,n$ e $i\neq j$.



### Interpretao dos parmetros

- $\beta_0$: valor esperado de $Y$ quando $X$  zero.

- $\beta_1$: aumento esperado em $Y$ quando $X$  acrescido de uma unidade.

## Modelo de regresso linear mltipla


Motivao: Deseja-se construir um modelo para explicar


-  $Y_j$: valor de mercado de uma casa utilizando variveis explicativas
-  $X_{1j}$: rea
-  $X_{2j}$: localizao
-  $X_{3j}$: valor da casa no ano anterior
-  $X_{4j}$: qualidade da construo

Um possvel modelo linear (nos parmetros) seria:

$$\large\begin{array}{c} Y_j=\beta_0 + \beta_1 X_{1j} + \beta_2 X_{2j} + \beta_3 X_{3j}+\beta_4 X_{4j} + \epsilon_j.\\
\end{array}
$$




Nomenclatura:
-  $Y_j$: varivel resposta (dependente),
-  $\beta_i$: parmetros desconhecidos,
-  $X_{ij}$: variveis explicativas (covariveis, variveis independentes),
-  $\epsilon_j$: erro aleatrio.


Suposies:

-  $E(\epsilon_j)=0$ para $ j=1,\ldots,n$,
-  $Var(\epsilon_j) = \sigma^2$ para $ j=1,\ldots,n$,
-  $Cov(\epsilon_i,\epsilon_j)=0$ para $ i,j=1,\ldots,n$ e $i\neq j$.






Poderamos estender esse modelo para $p$ covariveis,


$$\large\begin{array}{c} Y_i=\beta_0 + \beta_1 X_{1i} + \beta_2 X_{2i} + \ldots +\beta_p X_{pi} + \epsilon_i, i=1,\ldots,n.\\  
\end{array}
$$

Note que a varivel resposta $Y_i$  unidimensional.



Poderamos "empilhar" os dados de $n$ indivduos em linhas. Teramos ento matricialmente  

$$Y=\left[\begin{array}{c}Y_1\\ Y_2\\ \vdots \\ Y_n\end{array}\right], \
X = \left[\begin{array}{cccc}
1 & X_{11} & \ldots & X_{1p} \\
1 & X_{21} & \ldots & X_{2p} \\
\vdots & \vdots & \ddots & \vdots\\
1 & X_{n1} & \ldots & X_{np} \\
\end{array}\right], \beta = \left[\begin{array}{c}\beta_0\\ \beta_1\\\vdots \\ \beta_p\end{array}\right], \  \epsilon = \left[\begin{array}{c}\epsilon_1\\ \epsilon_2\\\vdots \\ \epsilon_n\end{array}\right]
$$  

ou seja,

$$Y_{n\times 1} = X_{n\times (p+1)} \beta_{(p+ 1)\times 1} + \epsilon_{n\times 1}. $$


### Interpretao dos parmetros

- $\beta_0$: valor esperado de $Y$ quando $X_{1i}, X_{2i}, \ldots, X_{pi}$ so todas zero.

- $\beta_k$: aumento esperado em $Y$ quando $X_k$  acrescido de uma unidade e todas as outras so mantidas fixadas, $k=1,\ldots,p$.

## Estimao dos parmetros

Alguns mtodos podem ser usados para estimar os parmetros, por exemplo

- Mtodo de mnimos quadrados ordinrios (EMQ ou MQO)

- Mtodo de mxima verossimilhana (EMV)


No modelo linear geral
$$\large Y = X \beta + \epsilon. $$

com as suposies


- $E( \epsilon) = {0}$,
- $Var( \epsilon) = \sigma^2 I$,


o **estimador de mnimos quadrados** que minimiza a soma de quadrados dos resduos,  dado por

$$\widehat{\beta} = (X ^{\top} X)^{-1} X ^{\top} Y.$$



Se $\epsilon\sim N({0},\sigma^2 I)$, ento

o **estimador de mxima verossimilhana** de $\beta$  dado (tambm) por

$$\widehat{\beta} = (X^{\top} X)^{-1} X^\top Y.$$

Nesse caso,

$$\widehat{\beta} \sim N\left(\beta, \sigma^2 (X^\top X)^{-1}\right)$$

e  comum estimar $\sigma^2$ com


$$\widehat{\sigma}^2 = MSE.$$

                 
            
            
**Observao**            

O EMV de $\beta$  no-viesado e consistente, que so boas propriedades estatsticas.


## Valor ajustado de $Y$            

O valor ajustado de $Y$, para um determinado $X = x$  obtido fazendo


$$\large \hat{Y} = X {\hat{\beta}}.$$





O erro quadrtico mdio, MSE,  usado para estimar $\sigma^2$, fazendo


$$\large \hat{\sigma}^2 = \displaystyle{\frac{SQE}{n-p}} =  \displaystyle{\frac{\displaystyle\sum_{i=1}^{n}(Y-\hat{Y})^\top(Y-\hat{Y})  }{n-p}} $$


            

## Coeficiente de determinao do modelo

O coeficiente de determinao, ou coeficiente de explicao do modelo,  dado por


$$R^2 = 1 - \displaystyle {\frac{SQE}{SQT}},$$

em que $SQT = Y^\top Y  -  \displaystyle\frac{1}{n}Y^\top \mathbb{1}^\top \mathbb{1} Y$, em que $\mathbb{1}$ indica um vetor de uns de mesma dimenso de $Y$.


Para levar em conta o aumento da explicao da variabilidade da resposta quando aumentamos o nmero de covariveis,  comum considerar o coeficiente de determinao do modelo ajustado:


$$R^2_{ajustado} = 1 - \displaystyle {\frac{n-1}{n-p}\frac{SQE}{SQT}}.$$


Tanto $R^2$ quanto $R^2_{ajustado}$ esto entre 0 e 1, e pode ser usado como um **indcio** de qualidade do ajuste, quanto maior o coeficiente de determinao, melhor  o modelo linear.




## Modelo de regresso linear multivariada


Considere agora que, para cada indivduo, sejam observadas $m$ variveis respostas, e que cada uma delas tenha uma relao linear com as p covariveis.


Assim, teriamos $m$ modelos de regresso:


$$\begin{array}{c} Y_1=\beta_{01} + \beta_{11} X_1 + \beta_{21} X_2 + \beta_{31} X_3+\ldots+\beta_{p1} X_p+\epsilon_{1}\\
 Y_2=\beta_{02} + \beta_{12} X_1 + \beta_{22} X_2 + \beta_{32} X_3+\ldots+\beta_{p2} X_p+\epsilon_{2}\\
 \vdots\\
 Y_m=\beta_{0m} + \beta_{1m} X_1 + \beta_{2m} X_2 + \beta_{3m} X_3+\ldots+\beta_{pm} X_p+\epsilon_{m}\\
\end{array}
$$


Para cada um dos $n$ indivduos, vamos observar as $m$ variveis resposta e as $p$ covariveis. Assim, podemos definir um modelo de regresso multivariado

$$Y_{n\times m} = X_{n\times (p+1)} \beta_{(p+1)\times m} + \epsilon_{n\times m}$$

em que



$$Y=\left[\begin{array}{cccc}
Y_{11} & Y_{12} & \ldots & Y_{1m} \\
Y_{21} & Y_{22} & \ldots & Y_{2m} \\
\vdots & \vdots & \ddots & \vdots\\
Y_{n1} & Y_{n2} & \ldots & Y_{nm} \\
\end{array}\right],
\
X = \left[\begin{array}{cccc}
1 & X_{11} & \ldots & X_{1p} \\
1 & X_{21} & \ldots & X_{2p} \\
\vdots & \vdots & \ddots & \vdots\\
1 & X_{n1} & \ldots & X_{np} \\
\end{array}\right],$$


$$\beta = \left[\begin{array}{cccc}
\beta_{01} & \beta_{02} & \ldots & \beta_{0m}\\
\beta_{11} & \beta_{12} & \ldots & \beta_{1m}\\
\vdots & \vdots & \ddots & \vdots\\
\beta_{p1} & \beta_{p2} & \ldots & \beta_{pm}\\
\end{array}\right], \  \epsilon = \left[\begin{array}{cccc}
\epsilon_{11} & \epsilon_{12} & \ldots & \epsilon_{1m} \\
\epsilon_{21} & \epsilon_{22} & \ldots & \epsilon_{2m} \\
\vdots & \vdots & \ddots & \vdots\\
\epsilon_{n1} & \epsilon_{n2} & \ldots & \epsilon_{nm} \end{array}\right]
$$

## Anlise de diagnstico

Os resduos contm indicativos de adequabilidade das suposies do modelo


Os **resduos ordinrios** do modelo so dados por

$$\large e =  Y-\widehat{Y}.$$

 comum construir grficos dos resduos ordinrios contra a ordem das observaes, os valores ajustados $\widehat{Y}$ e $X_i$, para algumas das variveis preditoras de interesse.


Espera-se que os resduos sejam aleatoriamente distribudos em torno de zero.







Alguns padres de resduos so ilustrados a seguir:


residuos3.pngresiduos2.pngresiduos1.png


Existem algumas propostas para a padronizao de resduos:


Temos que

$$e =  Y-\widehat{Y}  =  Y -  X  (X ^{\top} X)^{-1} X ^{\top} Y = (I - X  (X ^{\top} X)^{-1} X ^{\top}) Y = (I-H) Y,$$

em que $H = X  (X ^{\top} X)^{-1} X ^{\top}$  a matriz hat (matriz chapu). Seja $h_{ii}$ o $i$-simo elemento da diagonal de $H$.

Pode-se mostrar que

$$Var(e_i) = (1-h_{ii})\sigma^2$$

Assim, podemos definir dois novos resduos:


- **Resduo Studentizado internamente:**

$$s_i = \displaystyle\frac{e_i}{s\sqrt{1-h_{ii}}},$$

em que $s$  uma estimativa de $\sigma$, usualmente a raiz do MSE.


- **Resduo Studentizado externamente**

$$t_i = \displaystyle\frac{e_i}{s(i)\sqrt{1-h_{ii}}},$$

em que $s(i)$  uma estimativa para $\sigma^2$ sem a observao i.


Observao: Com $h_{ii}$  possvel identificar **pontos de alavanca**, que so outliers no espao dos X e no necessariamente so ponto influentes, ou seja, no necessariamente mudam a inferncia do modelo.

Existem tcnicas para identificar **pontos influentes**, como DFFITS e Distncia de Cook, que veremos na prtica.

```python
# Grficos de diagnstico para o modelo linear simples com estimadores MQO

# Fonte: https://stackoverflow.com/questions/46607831/python-linear-regression-diagnostic-plots-similar-to-r
# Fonte: https://stackoverflow.com/questions/46304514/access-standardized-residuals-cooks-values-hatvalues-leverage-etc-easily-i/55764402#55764402

from matplotlib import pyplot as plt
from pandas.core.frame import DataFrame
import scipy.stats as stats
import statsmodels.api as sm


def linear_regression(df: DataFrame) -> DataFrame:
    """Perform a univariate regression and store results in a new data frame.

    Args:
        df (DataFrame): orginal data set with x and y.

    Returns:
        DataFrame: another dataframe with raw data and results.
    """
    mod = sm.OLS(endog=df['y'], exog=df['x']).fit()
    influence = mod.get_influence()

    res = df.copy()
    res['resid'] = mod.resid
    res['fittedvalues'] = mod.fittedvalues
    res['resid_std'] = mod.resid_pearson
    res['leverage'] = influence.hat_matrix_diag
    return res


def plot_diagnosis(df: DataFrame):
    fig, axes = plt.subplots(nrows=2, ncols=2)
    plt.style.use('seaborn')

    # Residual against fitted values.
    df.plot.scatter(
        x='fittedvalues', y='resid', ax=axes[0, 0]
    )
    axes[0, 0].axhline(y=0, color='grey', linestyle='dashed')
    axes[0, 0].set_xlabel('Fitted Values')
    axes[0, 0].set_ylabel('Residuals')
    axes[0, 0].set_title('Residuals vs Fitted')

    # qqplot
    sm.qqplot(
        df['resid'], dist=stats.t, fit=True, line='45',
        ax=axes[0, 1], c='#4C72B0'
    )
    axes[0, 1].set_title('Normal Q-Q')

    # The scale-location plot.
    df.plot.scatter(
        x='fittedvalues', y='resid_std', ax=axes[1, 0]
    )
    axes[1, 0].axhline(y=0, color='grey', linestyle='dashed')
    axes[1, 0].set_xlabel('Fitted values')
    axes[1, 0].set_ylabel('Sqrt(|standardized residuals|)')
    axes[1, 0].set_title('Scale-Location')

    # Standardized residuals vs. leverage
    df.plot.scatter(
        x='leverage', y='resid_std', ax=axes[1, 1]
    )
    axes[1, 1].axhline(y=0, color='grey', linestyle='dashed')
    axes[1, 1].set_xlabel('Leverage')
    axes[1, 1].set_ylabel('Sqrt(|standardized residuals|)')
    axes[1, 1].set_title('Residuals vs Leverage')

    plt.tight_layout()
    plt.show()
```

```python
df = data=anscombe.query("dataset == 'IV'")
df = df.drop('dataset', axis=1)
```

## Mtricas para avaliar a qualidade do ajuste

### Erro absoluto mdio (EAM, MAE)

O erro quadrtico mdio (EQM) ou mean absolute error (MAE)  a mdia do valor absoluto dos erros.


$\mbox{MAE} = \displaystyle\frac{1}{n} \displaystyle\sum_{j=1}^{n}|y_j - \widehat{y}_j|$



### Erro quadrtico mdio (EQM, MSE)

O erro quadrtico mdio (EQM) ou mean squared error (MSE)   a mdia dos erros quadrticos.


$\mbox{MSE} = \displaystyle\frac{1}{n} \displaystyle\sum_{j=1}^{n}(y_j - \widehat{y}_j)^2$




### Raiz do erro quadrtico mdio (REQM, RMSE)

A raiz do erro quadrtico mdio (REQM) ou root mean squared error (RMSE)  a raiz da mdia dos erros quadrticos.




$\mbox{RMSE} = \displaystyle\sqrt{\frac{1}{n} \displaystyle\sum_{j=1}^{n}(y_j - \widehat{y}_j)^2}$

```python
# R2 ajustado de
# https://www.kaggle.com/burhanykiyakoglu/predicting-house-prices

def adjustedR2(r2,n,k):
    return r2-(k-1)/(n-k)*(1-r2)
```

```python
# Fonte: https://www.kaggle.com/burhanykiyakoglu/predicting-house-prices

train_data,test_data = train_test_split(df,train_size = 0.8,random_state=3)

features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront', 'view', 'condition', 'grade']
complex_model_1 = linear_model.LinearRegression()
complex_model_1.fit(train_data[features],train_data['price'])

print('Intercept: {}'.format(complex_model_1.intercept_))
print('Coefficients: {}'.format(complex_model_1.coef_))

pred = complex_model_1.predict(test_data[features])
rmsecm = float(format(np.sqrt(metrics.mean_squared_error(test_data['price'],pred)),'.3f'))
rtrcm = float(format(complex_model_1.score(train_data[features],train_data['price']),'.3f'))

artrcm = float(format(adjustedR2(complex_model_1.score(train_data[features],train_data['price']),train_data.shape[0],len(features)),'.3f'))
rtecm = float(format(complex_model_1.score(test_data[features],test_data['price']),'.3f'))
artecm = float(format(adjustedR2(complex_model_1.score(test_data[features],test_data['price']),test_data.shape[0],len(features)),'.3f'))
cv = float(format(cross_val_score(complex_model_1,df[features],df['price'],cv=5).mean(),'.3f'))

r = evaluation.shape[0]
evaluation.loc[r] = ['Multiple Regression-1','selected features',rmsecm,rtrcm,artrcm,rtecm,artecm,cv]
evaluation.sort_values(by = '5-Fold Cross Validation', ascending=False)
```

```python
# Visualizao do modelo de regresso mltipla em Python

## Fonte: https://aegis4048.github.io/mutiple_linear_regression_and_visualization_in_python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from mpl_toolkits.mplot3d import Axes3D

######################################## Preparao dos dados #########################################


X = df[['sqft_living', 'grade']].values.reshape(-1,2)
Y = df['price_transformado']

######################## Preparao para a visualizao ###############################

x = X[:, 0]
y = X[:, 1]
z = Y

x_pred = np.linspace(290, 13540, 30)   # grade de valores para x
y_pred = np.linspace(1, 13, 30)  # grade de valores para y
xx_pred, yy_pred = np.meshgrid(x_pred, y_pred)
model_viz = np.array([xx_pred.flatten(), yy_pred.flatten()]).T

################################################ Treinamento do modelo #############################################

ols = linear_model.LinearRegression()
model = ols.fit(X, Y)
predicted = model.predict(model_viz)

############################################## Avaliao ############################################

r2 = model.score(X, Y)

############################################## Plota o grfico ################################################

plt.style.use('default')

fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(131, projection='3d')
ax2 = fig.add_subplot(132, projection='3d')


axes = [ax1, ax2]

for ax in axes:
    ax.plot(x, y, z, color='k', zorder=15, linestyle='none', marker='o', alpha=0.5)
    ax.scatter(xx_pred.flatten(), yy_pred.flatten(), predicted, facecolor=(0,0,0,0), s=20, edgecolor='#70b3f0')
    ax.set_xlabel('sqft_living', fontsize=12)
    ax.set_ylabel('grade', fontsize=12)
    ax.set_zlabel('price_transformado', fontsize=12)
    ax.locator_params(nbins=4, axis='x')
    ax.locator_params(nbins=5, axis='x')


ax1.view_init(elev=20, azim=-10)
ax2.view_init(elev=0, azim=290)


fig.suptitle('$R^2 = %.2f$' % r2, fontsize=20)

fig.tight_layout()
```
