# Modelos lineares


**Objetivos**

Predizer $Y$ a partir do conhecimento de variveis preditoras em $X = x$.


Um modelo linear considerando pares observados de $(X_i, Y_i), i=1,\ldots,n$, em que $X_i$  um vetor de preditoras  dado por

$$\large{Y_i = X_i^\top\beta+\epsilon_i,}$$

em que, para o $i$-simo elemento amostral, temos

- $Y_i$  a **varivel resposta** (aleatria observvel),

- $X_i$ contm **variveis preditoras** (vetor conhecido, ou seja, no-aleatrio),

- $\beta$  um **vetor de parmetros de interesse**, que queremos estimar,

- $\epsilon_i$  o **erro aleatrio** (no observvel).


Suposies do modelo linear geral:

- $ E(Y_i) = \mu_i =  X_i^\top\beta$

- $\epsilon_i \stackrel{i.i.d}{\sim} N(0, \sigma^2)$

- Consequentemente $ Y_i|X_i \sim N(\mu_i, \sigma^2).$

O modelo linear (normal)  muito til, mas nem sempre as suposies esto satisfeitas.  importante ter metodologias mais adequadas, em especial, para casos em que

- A varivel resposta no possui distribuio Normal.

- A associao entre a resposta e as preditoras no  linear.



#  Modelo linear generalizado (MLG)

Nelder, John; Wedderburn, Robert (1972). Generalized Linear Models. Blackwell Publishing. Journal of the Royal Statistical Society. Series A (General). 135: 370384. JSTOR 2344614. doi:10.2307/2344614

Os MLGs propem a modelagem para variveis na famlia exponencial de distribuies, que inclui

- Binomial (Bernoulli)
- Poisson
- Normal
- Binomial Negativa
- Gama
- Gaussiana Inversa
- outras



### A famlia exponencial de distribuies

Seja $Y_i$ uma varivel aleatria cuja distribuio pode ser escrita na forma


$$\large f(y,\theta_i, \phi) = exp[\phi\{y \theta_i - b(\theta_i)\} + c(y, \phi)]$$


Dizemos que a distribuio de $Y_i$ pertence  famlia exponencial e escrevemos $$\large Y_i \sim FE(\mu_i, \phi),$$


em que $\mu_i = E(Y_i) = b'(\theta_i)$  o parmetro de posio e $\phi^{-1}$ o parmetro de disperso.


Alm disso, $Var(Y_i) = \phi^{-1}V_i$ com $V_i = \displaystyle\frac{d\mu}{d\theta}$ a **funo de varincia**.



Ento dizemos que um MLG  determinado pela **funo de ligao**
$g(\mu_i) = \eta_i$ e pela **funo de varincia** $V(\mu)$.

## Funo de ligao

O modelo linear generalizado  definido por

$$\begin{array}{c}\large Y_i \sim FE(\mu_i, \phi) \\\large g(\mu_i) = \eta_i\end{array}$$

em que

- $ \eta_i = X_i^\top\beta$  o **preditor linear**,

- $\beta = (\beta_0, \beta_1,\ldots, \beta_p)^\top$, $p<n$  um vetor de parmetros desconhecidos (coeficientes da regresso),

- $ X_i = (1, X_{i1},\ldots, X_{ip})^\top$ representa os valores de $p$ variveis preditoras e

- $g(\mu_i)$  a **funo de ligao**, uma funo montona e diferencivel.






### Exemplos


#### Modelo Normal

Se $Y\sim N(\mu, \sigma^2)$, com densidade

$$\begin{array}{lll}f(y) &=& \displaystyle\frac{1}{\sqrt{2\pi\sigma^2}}\exp\left\{-\frac{1}{2\sigma^2}(y-\mu)^2\right\} = \\ &=&\displaystyle\exp\left[ \left\{ \frac{1}{\sigma^2}\left( \mu y - \frac{\mu^2}{2}\right)  \right\}- \frac{1}{2} \left\{ \log 2 \pi \sigma^2 + \frac{y^2}{\sigma^2}\right\}\right],\end{array}$$

em que $-\infty < \mu < \infty$ e $\sigma^2>0$.

Logo, $\theta = \mu$, $b(\theta) = \theta^2/2 $, $\phi=\sigma^{-2}$ e $c(y, \phi) = \frac{1}{2} \log(\phi/2\pi) - \phi y^2/2$. Verifica-se que $V(\mu) = 1$.



#### Modelo Poisson: para dados de contagem

Se $Y\sim P(\mu)$, com densidade dada por

$$ f(y) =P(Y=y) =   \displaystyle\frac{e^{-\mu} \mu^y}{y!} = exp(y \log\mu -\mu -\log y!), $$

em que $\mu>0$ e $y=0,1,...$

Assim $\theta = \log\mu$, $b(\theta) = e^\theta$, $\phi=1$ e $c(y, \phi) = -\log y!$. Segue que $V(\mu) = \mu$.



#### Modelo Binomial: para modelar propores


Se $Y^\star$  a proporo de sucessos em $n$ ensaios independentes de Bernoulli, cada um com probabilidade de sucesso $\mu$, ento $nY^\star\sim B(n,\mu)$, com densidade dada por

$$\begin{array}{lll}f(y)  &=& P(Y=y) = \left(\begin{array}{c} n\\ ny^\star\end{array}\right)\mu ^{ny^\star}(1-\mu)^{n-ny^\star} = \\ &=&\exp\left\{\log  \left(\begin{array}{c} n\\ ny^\star\end{array}\right) + ny^\star \log\left(\displaystyle\frac{\mu}{1-\mu} + n\log(1-\mu)\right)\right\},\end{array}  $$

em que $\mu>0$, $y^\star<1$. Nesse caso, $\phi =n$, $\theta = \log\displaystyle\left(\frac{\mu}{1-\mu}\right)$, $b(\theta) = \log(1+e^\theta)$, e $c(y^\star, \phi) = \log  \left(\begin{array}{c} \phi\\ \phi y^\star\end{array}\right)$. Segue que $V(\mu) = \mu(1-\mu)$.




### Funo de ligao cannica

Se $\theta_i = \eta_i = X_i^\top\beta$, ento $\eta$  chamada de **ligao cannica**.

- No modelo **normal**, a ligao cannica  $\eta = \mu $

- No modelo **Poisson**, a ligao cannica  $\eta = \log\mu$

- No modelo **binomial**, a ligao cannica  $\eta = \log\displaystyle\left\{\frac{\mu}{1-\mu}\right\}$



## O modelo de regresso logstica

Um dos modelos lineares mais utilizados  o modelo de regresso logstica.

Neste modelo, consideramos


$Y_i = \left\{\begin{array}{l}1, \mbox{ se o indivduo i possui determinada caracterstica}\\ 0, \mbox{ caso contrrio}\end{array}\right.$

Supondo que $P(Y_i|X_i) = \pi(X_i)$ e que

$$\log\left\{\displaystyle\frac{\pi(x)}{1-\pi(x)}\right\} = \alpha+\beta x$$

e queremos estimar os parmetros $\alpha$ e $\beta$ para compreender como $X_i$ e $Y_i$ esto associados.



Suponha que $X_i=1$ indique que o indivduo $i$ possui um fator de risco para uma determinada doena (indicada por $Y_i=1)$ e foram observados $n_1$ indivduos com a presena deste fator, e $n_2$ indivduos sem a presena deste fator ($X_i=0$).


Para os indivduos que possuiam o fator, a **chance** de desenvolvimento da doena fica

$$\displaystyle\frac{\pi(1)}{1-\pi(1)} = e^{\alpha + \beta},$$

enquanto que a **chance** de desenvolvimento da doena no indivduo com ausncia do fator 

$$\displaystyle\frac{\pi(0)}{1-\pi(0)} = e^{\alpha }.$$


A razo de chances nesse caso fica

$$\psi =  \displaystyle\frac{\pi(1)/(1-\pi(1))}{\pi(0)/(1-\pi(0))} = e^\beta.$$



### Regresso logstica mltipla


O modelo de regresso logstica pode ser estendido para incluir $p$ variveis preditoras:

$Y_i = \left\{\begin{array}{l}1, \mbox{ se o indivduo i possui determinada caracterstica}\\ 0, \mbox{ caso contrrio}\end{array}\right.$

Supondo que $P(Y_i|X_i) = \pi(X_i)$ e que

$$\log\left\{\displaystyle\frac{\pi(x)}{1-\pi(x)}\right\} = \beta_0 + \beta_1 x_{1} + \beta_2 x_{2} + \ldots + \beta_p x_{p}.$$


## A estimao dos parmetros

No **enfoque clssico**,  comum obter os **estimadores de mxima verossimilhana** por mtodos iterativos como

- Mtodo de Newton-Raphson

- Mtodo escore de Fisher

- Mtodo dos mnimos quadrados ou mnimos quadrados reponderados


No **enfoque Bayesiano**, supomos distribuies de probabilidade para os coeficientes da regresso com mtodos computacionais e obtemos estimativas a partir de amostras da distribuio a posteriori dos parmetros, como Monte Carlo Hamiltoniano ou No U-Turn Sampler (NUTS).

```python
import pandas as pd

# Dados banco - Leitura dos dados
dados = pd.read_csv('https://raw.githubusercontent.com/cibelerusso/Estatistica-Ciencia-Dados/main/Data/dados_banco.csv', index_col=0)

# Vamos trabalhar com uma amostra como na Aula 6
dados = dados.sample(n=500, replace=False, random_state=10)

dados.head()
```

```python
### Ajustando um MLG com resposta binria, iniciando com 4 preditoras

preditoras = dados_treino[['Idade','Devedor_cartao','Salario','Saldo_cc']]
resposta = dados_treino[['Inadimplente']]
```

```python
X_teste = dados_teste[['Idade','Devedor_cartao','Salario','Saldo_cc']]
Y_teste = dados_teste[['Inadimplente']]

predito = res.predict(X_teste)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_squared_error(predito, Y_teste)
```

```python
res.aic
```

## Modelos lineares generalizados: um enfoque Bayesiano



```python
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import patsy as pt
import pymc as pm
import seaborn as sns

print(f"Running on PyMC3 v{pm.__version__}")
import warnings

# Ignorar avisos especficos do Seaborn
warnings.filterwarnings("ignore", module="seaborn")
```

```python
%config InlineBackend.figure_format = 'retina'
RANDOM_SEED = 8927
np.random.seed(RANDOM_SEED)
az.style.use("arviz-darkgrid")
```

### Aplicao

Fonte: https://www.pymc.io/projects/examples/en/2022.01.0/generalized_linear_models/GLM-negative-binomial-regression.html

Este conjunto de dados fictcio foi criado para emular alguns dados criados como parte de um estudo do self quantificado, e os dados reais so mais complicados do que isso.

Descrio dos dados

- O sujeito espirra N vezes por dia, registrado como nsneeze (int)

- O sujeito pode ou no beber lcool durante o dia, registrado como lcool (booleano)

- O sujeito pode ou no tomar um medicamento anti-histamnico durante esse dia, registrado como ao negativa denominada (booleano)

- Os dados so agregados por dia, para produzir uma contagem total de espirros naquele dia, com um sinalizador booleano para o uso de lcool e anti-histamnicos, com a grande suposio de que os espirros tm uma relao causal direta.

- Crie 4000 dias de dados: contagens dirias de espirros que so distribudos por Poisson durante o consumo de lcool e o uso de anti-histamnicos

```python
# decide poisson theta values
theta_noalcohol_meds = 1  # no alcohol, took an antihist
theta_alcohol_meds = 3  # alcohol, took an antihist
theta_noalcohol_antihist = 6  # no alcohol, no antihist
theta_alcohol_antihist = 36  # alcohol, no antihist

# create samples
q = 1000
df = pd.DataFrame(
    {
        "nsneeze": np.concatenate(
            (
                np.random.poisson(theta_noalcohol_meds, q),
                np.random.poisson(theta_alcohol_meds, q),
                np.random.poisson(theta_noalcohol_antihist, q),
                np.random.poisson(theta_alcohol_antihist, q),
            )
        ),
        "alcohol": np.concatenate(
            (
                np.repeat(False, q),
                np.repeat(True, q),
                np.repeat(False, q),
                np.repeat(True, q),
            )
        ),
        "antihist": np.concatenate(
            (
                np.repeat(False, q),
                np.repeat(False, q),
                np.repeat(True, q),
                np.repeat(True, q),
            )
        ),
    }
)
```

```python
df.tail()
```

```python
df.groupby(["alcohol", "antihist"]).mean().unstack()
```

```python
g = sns.catplot(
    x="nsneeze",
    row="antihist",
    col="alcohol",
    data=df,
    kind="count",
    height=4,
    aspect=1.5,
)
```

```python
fml = "nsneeze ~ alcohol + antihist + alcohol:antihist"  # full patsy formulation
```

```python
fml = "nsneeze ~ alcohol * antihist"  # lazy, alternative patsy formulation
```

```python
(mx_en, mx_ex) = pt.dmatrices(fml, df, return_type="dataframe", NA_action="raise")
```

```python
pd.concat((mx_ex.head(3), mx_ex.tail(3)))
```

```python
with pm.Model() as mdl_Poisson:

    # define priors, weakly informative Normal
    b0 = pm.Normal("b0_intercept", mu=0, sigma=10)
    b1 = pm.Normal("b1_alcohol[T.True]", mu=0, sigma=10)
    b2 = pm.Normal("b2_antihist[T.True]", mu=0, sigma=10)
    b3 = pm.Normal("b3_alcohol[T.True]:antihist[T.True]", mu=0, sigma=10)

    # define linear model and exp link function
    eta = (
        b0
        + b1 * mx_ex["alcohol[T.True]"]
        + b2 * mx_ex["antihist[T.True]"]
        + b3 * mx_ex["alcohol[T.True]:antihist[T.True]"]
    )

    ## Define Poisson likelihood
    y = pm.Poisson("y", mu=np.exp(eta), observed=mx_en["nsneeze"].values)
```

**Obteno das amostras**

```python
with mdl_Poisson:
    inf_Poisson = pm.sample(1000, tune=1000, cores=4, return_inferencedata=True)
```

**Diagnstico do modelo**

```python
az.plot_trace(inf_Poisson)
```

**Estimao dos parmetros**

```python
np.exp(az.summary(inf_Poisson)[["mean", "hdi_3%", "hdi_97%"]])
```
