# Questão 1

Uma máquina que enche potes de iogurte de forma automática em uma indústria de laticínios está regulada de modo a servir uma média de 100ml por recipiente. Se a quantidade servida por embalagem seguir uma distribuição normal com desvio padrão de 10 ml, determine a proporção de embalagens que transbordarão se forem usadas embalagens de 110 ml. (Aproxime para duas casas decimais).

```python
sigma = 10
mu = 100
capacidade_embalagem = 110
# P(X >= 110)
```

```python
probabilidade_transbordamento = 1 - norm.cdf(capacidade_embalagem, mu, sigma)

print(f"A proporção de embalagens que transbordarão é {probabilidade_transbordamento:.2f}")
```

A função `norm.cdf` da biblioteca `scipy.stats` é usada para calcular a **função de distribuição acumulada** (CDF) para uma distribuição normal. Essa função é essencial para encontrar a probabilidade de que uma variável aleatória normalmente distribuída seja menor ou igual a um determinado valor.

Aqui está o que cada parte da função `norm.cdf(capacidade_embalagem, mu, sigma)` significa:
- `norm.cdf`: Chama a função CDF para a distribuição normal.
- `capacidade_embalagem`: É o valor limite que estamos interessados, neste caso, 110 ml. Estamos calculando a probabilidade de que a variável aleatória seja menor ou igual a este valor.
- `mu`: É a média da distribuição, que foi dada como 100 ml.
- `sigma`: É o desvio padrão da distribuição, que foi dado como 10 ml.

Quando você executa `norm.cdf(capacidade_embalagem, mu, sigma)`, está calculando a probabilidade de que a quantidade servida seja **menor ou igual a 110 ml**. Para encontrar a proporção de embalagens que **transbordarão**, você subtrai esse valor de 1, porque estamos interessados na parte da distribuição que está **acima de 110 ml**.

Portanto, `1 - norm.cdf(capacidade_embalagem, mu, sigma)` dá a probabilidade de que a quantidade servida seja **maior que 110 ml**, o que corresponde à proporção de embalagens que transbordarão.

Espero que isso esclareça a função para você! Se tiver mais perguntas ou precisar de mais ajuda, estou à disposição.

```python
# Definir média e desvio padrão
md = 100
dp = 10

# Gerar valores para o eixo x
x = np.linspace(md - 4 * dp, md + 4 * dp, 1000)

# Calcular os valores da distribuição normal padrão para cada valor em x
y = st.norm.pdf(x, md, dp)

# Plotar a curva da distribuição normal
plt.plot(x, y, label='Distribuição Normal')
plt.xlabel('Volume (ml)')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição Normal - Média=100, Desvio Padrão=10')

# Pintar a área onde x < 110
plt.fill_between(x, y, where=(x <= 110), color='lightblue', alpha=0.5)
plt.fill_between(x, y, where=(x > 110), color='red', alpha=0.5)

plt.legend()
plt.grid(True)
plt.show()
```

# Questão 2
A altura de alunos em uma escola de São Paulo pode ser modelada como uma distribuição normal com média 1,70m e desvio padrão 0,5m. Determine a probabilidade de que um aluno selecionado de forma aleatória tenha altura menor do que 1,80m.

```python
media = 1.70
desvio_padrao = 0.5
limite_inferior = 1.8
```

```python
1 - norm.cdf(limite_inferior, media, desvio_padrao)
```

```python
from scipy.stats import norm

# Média (mu) e desvio padrão (sigma)
mu = 1.70
sigma = 0.5

# Altura para calcular a probabilidade
altura = 1.80

# Calculando a probabilidade de que a altura seja menor que 1,80m
probabilidade = norm.cdf(altura, mu, sigma)

print(f"A probabilidade de que um aluno tenha altura menor do que 1,80m é {probabilidade:.2f}")
```
