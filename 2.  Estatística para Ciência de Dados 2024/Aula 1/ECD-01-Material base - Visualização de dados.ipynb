## Anlise exploratria de dados


**Anlise descritiva** ou **anlise exploratria de dados** (AED) tem como objetivos bsicos:


- explorar os dados para descobrir ou identificar aspectos ou padres de maior interesse,


- representar os dados de forma a destacar ou chamar a ateno para aspectos ou padres que podem ou no se confirmar inferencialmente.



Tukey (1977) chama a **anlise exploratria de dados** de **trabalho de detetive**, que busca pistas e evidncia, e a **anlise confirmatria de dados**  um **trabalho judicial ou quase-judicial**, que analisa e avalia a fora das provas e da evidncia.




Tukey tambm diz que: **"A anlise exploratria de dados nunca conta a histria toda, mas nada  to perfeito para ser considerado a pedra fundamental, um primeiro passo para a anlise de dados"**.




 importante salientar que a AED  um trabalho inicial, a pedra fundamental, e os resultados devem ser analisados com uma anlise confirmatria.

## Medidas-resumo

- **Medidas de posio**
    - **Mdia:** boas propriedades estatsticas
    - **Mediana:** medida resistente
    - **Moda:** valor mais frequente

- **Medidas de disperso**
    - **Desvio-padro**

- **Assimetria**: Assimetria da distribuio dos dados


- **Curtose**: Achatamento da distribuio


- **Medidas de associao**: Covarincia, Coeficiente de correlao de Pearson, Coeficiente de correlao de Spearman

### Medidas de posio

Daqui em diante, vamos estabelecer $X_1,\ldots, X_n$  uma amostra aleatria e $x_1,\ldots, x_n$ os **dados observados** dessa amostra. As medidas aqui apresentadas so **amostrais** e so obtidas a partir de $x_1,\ldots, x_n$.

A **mdia** (amostral observada)  definida como


$\bar{x} = \displaystyle{\frac{\displaystyle\sum_{i=1}^{n} x_i}{n}}$


Considere agora os dados ordenados $x_{(1)},\ldots, x_{(n)}$, isto ,

$x_{(1)} = min(x_1,\ldots, x_n)$ e $x_{(n)} = max(x_1,\ldots, x_n)$.

Se $n$  mpar, a posio central  $c = (n + 1) / 2$. Se $n$  par, as posies centrais so $c = n / 2$ e $c + 1 = n / 2  + 1$.


A **mediana**  definida como


$Md = \bigg \{\begin{array}{l}x_{(c)}, \mbox{se n  mpar}\\
\displaystyle\frac{x_{(c)}+x_{(c+1)}}{2}, \mbox{se n  par}\end{array}$



A **moda**  o valor mais frequente da amostra. No necessariamente existe.





Um **quantil**  o valor que provoca uma diviso conveniente nos valores ordenados. O quantil de 10% divide os dados de tal forma que 10% dos menores valores fiquem " sua esquerda". O quantil de 50%  a mediana.



Os **quartis** dividem os dados em pores de 25%.



Os **decis** dividem os dados em pores de 10%.



Os **percentis** dividem os dados em pores de 1%.



### Medidas de disperso

A **varincia amostral**  dada por

$s^2 = \displaystyle{\frac{\displaystyle\sum_{i=1}^{n}(x_i-\bar{x})^2}{n}}.$


O **desvio padro**  dado por

$s = \displaystyle\sqrt{\frac{\displaystyle\sum_{i=1}^{n}(x_i-\bar{x})^2}{n}}.$


 comum, entretanto, utilizar as medidas corrigidas:


**Varincia amostral corrigida:**

$s^2 = \displaystyle{\frac{\displaystyle\sum_{i=1}^{n}(x_i-\bar{x})^2}{n-1}}$


**Desvio padro corrigido:**

$s = \displaystyle\sqrt{\frac{\displaystyle\sum_{i=1}^{n}(x_i-\bar{x})^2}{n-1}}$





A **amplitude**  dada por

$A = x_{(n)} -x_{(1)}.$



O **coeficiente de variao** (amostral)  dado pela razo entre o desvio-padro e a mdia

$CV = \displaystyle{\frac{s}{\bar{x}}}$

### Assimetria

- Distribuio simtrica: mdia = mediana = moda
- Distribuio assimtrica  direita: moda < mediana < mdia
- Distribuio assimtrica  esquerda: mdia < mediana < moda


### Curtose

- Distribuies mesocrticas: achatamento da distribuio normal
- Distribuies leptocrticas: distribuio mais concentrada
- Distribuies platicrticas: distribuio mais achatada




```python
# Ilustrao das medidas mdia, moda, mediana para dados simtricos
# Adaptado de https://stackoverflow.com/questions/51417483/mean-median-mode-lines-showing-only-in-last-graph-in-seaborn/51417635

from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.DataFrame({"rating": [5, 6, 6, 7, 7, 7, 7, 8, 8, 9]})

f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)})
mean=df['rating'].mean()
median=df['rating'].median()
mode=df['rating'].mode().values[0]

sns.boxplot(data=df, x="rating", ax=ax_box)
ax_box.axvline(mean, color='r', linestyle='--')
ax_box.axvline(median, color='g', linestyle='-')
ax_box.axvline(mode, color='b', linestyle='-')

sns.histplot(data=df, x="rating", ax=ax_hist, kde=True)
ax_hist.axvline(mean, color='r', linestyle='--', label="Mean")
ax_hist.axvline(median, color='g', linestyle='-', label="Median")
ax_hist.axvline(mode, color='b', linestyle='-', label="Mode")

plt.legend()

ax_box.set(xlabel='')
plt.show()
```

```python
# Ilustrao das medidas mdia, moda, mediana para dados assimtricos  direita ou assimetria positiva
# Adaptado de https://stackoverflow.com/questions/51417483/mean-median-mode-lines-showing-only-in-last-graph-in-seaborn/51417635

from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import statistics

df = pd.DataFrame({"rating": [1,1,1,2,2,3,4,5,5,10]})

f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)})
mean=df['rating'].mean()
median=df['rating'].median()
mode=df['rating'].mode().values[0]

sns.boxplot(data=df, x="rating", ax=ax_box)
ax_box.axvline(mean, color='r', linestyle='--')
ax_box.axvline(median, color='g', linestyle='-')
ax_box.axvline(mode, color='b', linestyle='-')

sns.histplot(data=df, x="rating", ax=ax_hist, kde=True)
ax_hist.axvline(mean, color='r', linestyle='--', label="Mean")
ax_hist.axvline(median, color='g', linestyle='-', label="Median")
ax_hist.axvline(mode, color='b', linestyle='-', label="Mode")

plt.legend()


ax_box.set(xlabel='')
plt.show()
```

```python
# Ilustrao das medidas mdia, moda, mediana para dados assimtricos  esquerda ou com assimetria negativa
# Adaptado de: https://stackoverflow.com/questions/51417483/mean-median-mode-lines-showing-only-in-last-graph-in-seaborn/51417635


from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import statistics

df = pd.DataFrame({"rating": [1, 4, 6, 8, 8, 9, 10, 10, 10, 10]})

f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)})
mean=df['rating'].mean()
median=df['rating'].median()
mode = statistics.mode(df['rating'])

sns.boxplot(data=df, x="rating", ax=ax_box)
ax_box.axvline(mean, color='b', linestyle='--')
ax_box.axvline(median, color='r', linestyle='-')
ax_box.axvline(mode, color='g', linestyle='-')

sns.histplot(data=df, x="rating", ax=ax_hist, kde=True)
ax_hist.axvline(mean, color='b', linestyle='--', label="Mean")
ax_hist.axvline(median, color='r', linestyle='-', label="Median")
ax_hist.axvline(mode, color='g', linestyle='-', label="Mode")

plt.legend()

ax_box.set(xlabel='')
plt.show()
```

### Curtose

Referncias:
- https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html
- https://pt.wikipedia.org/wiki/Curtose

Medida que caracteriza o achatamento da curva.

- Curtose $\approx 0$: achatamento da curva normal
- Curtose $>0$: leptocrtica, distribuio mais afunilada
- Curtose $<0$: platicrtica, distribuio mais achatada

Obs: Distribuio normal https://www.spss-tutorials.com/normal-distribution/


```python
from scipy.stats import norm, kurtosis

data = norm.rvs(size=100000)

kurtosis(data)
```

```python
# Fonte: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html

import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import kurtosis
import numpy as np


x = np.linspace(-5, 5, 100)
ax = plt.subplot()
distnames = ['laplace', 'norm', 'uniform']

for distname in distnames:
    if distname == 'uniform':
        dist = getattr(stats, distname)(loc=-2, scale=4)
    else:
        dist = getattr(stats, distname)
    data = dist.rvs(size=1000)
    kur = kurtosis(data, fisher=True)
    y = dist.pdf(x)
    ax.plot(x, y, label="{}, {}".format(distname, round(kur, 3)))
    ax.legend()

# Normal: mesocrtica
# Laplace: leptocrtica
# Uniforme: platicrtica
```

## Medidas de associao entre variveis quantitativas


Sejam $X$ e $Y$ variveis quantitativas de interesse e as amostras aleatrias observadas $x_1,\ldots,x_n$ e $y_1,\ldots,y_n$, respectivamente. As medidas de associao mais utilizadas so:


### Covarincia (amostral)

$s_{XY} = \displaystyle{\frac{\displaystyle\sum_{i=1}^{n}(x_i-\bar{x})(y_i-\bar{y})}{n-1}}$




### Coeficiente de correlao linear (amostral) de Pearson

Referncia: https://pt.wikipedia.org/wiki/Coeficiente_de_correla%C3%A7%C3%A3o_de_Pearson


$r = \displaystyle{\frac{s_{XY}}{\sqrt{s^2_X s^2_Y }}}$


Propriedade:

 $-1 \leq r \leq 1$


 comum usar as seguintes classificaes:

1. $r=1$ indica uma correlao perfeita e positiva

2. $r=-1$ indica uma correlao perfeita e negativa

3. $0.7 \leq |r| \leq 1$ indica uma correlao forte

4. $0.5 \leq |r| \leq 0.69$ indica uma correlao moderada

5. $0 \leq |r| \leq 0.49$ indica uma correlao fraca




### Coeficiente de correlao de Spearman

Avalia relaes montonas entre duas variveis

Referncia: https://pt.wikipedia.org/wiki/Coeficiente_de_correla%C3%A7%C3%A3o_de_postos_de_Spearman

## Representao grfica e tabular de dados

- **Variveis qualitativas:**

    - **Tabela de frequncia:** Resume a informao dos dados de forma a possibilitar a observao de frequncias absolutas ou relativas de cada categoria das variveis qualitativas (ou valores assumidos pelas variveis quantitativas discretas).
    
    - **Grfico de barras** Representao grfica das frequncias de cada categoria das variveis qualitativas (ou valores assumidos pelas variveis quantitativas discretas). As barras so separadas.
    
    - **Grfico de Pareto**: Grfico de barras + frequncias acumuladas das categorias.
    
    - **Grfico de setores (pizza)**: Representao grfica das propores das categorias das variveis quantitativas discretas.
    
    
- **Variveis quantitativas discretas:**
    - **Tabelas de frequncias**
       
    - **Grficos de barras**
    
    - **Grficos de pontos**
    
    
    
- **Variveis quantitativas contnuas:**

    - **Histogramas**: Representao grfica para uma aproximao da distribuio de uma varivel quantitativa contnua, discretizada em classes de tamanhos convenientes. As barras so adjacentes. Permitem observar a localizao, disperso, assimetria, nmero de picos, curtose dos dados.
    
    - **Grficos de linhas (dados coletados ao longo do tempo)**
        
    - **Boxplots (grficos de caixas)**: Representao grfica inteligente que permite a observao da localizao, disperso, assimetria, pontos discrepantes (outliers).
