Dentre as strings resultantes, 'Quarter-finals' indica que a linha corresponde a uma partida de semi-final de uma copa.

Utilizando o método query, encontre as linhas contendo as partidas de quartas-de-final das copas e construa uma lista com os nomes dos times que disputaram as quartas-de-final da copa de 1994. 

Qual dos times abaixo não aparece na lista obtida:

a) Brazil<br>
**b) Czechoslovakia<br>**
c) Netherlands<br>
c) Bulgaria<br>

__DICA:__ Utilize o método query duas vezes, a primeira para obter um DataFrame com os times nas quartas-de-final e a segunda para obter um DataFrame com os times nas quartas-de-final da copa de 1994. Os nomes dos times nas quartas-de-final estão nas colunas `HomeTeamName` e `AwayTeamName`.

````python
copa_partidas.query("Year == 1994 and Stage == 'Quarter-finals'")[['HomeTeamName', 'AwayTeamName']]
````
